```python
import pandas as pd
import openpyxl
from openpyxl.styles import PatternFill

# 「2022_年間売上表.xlsx」と「2023_年間売上表.xlsx」のSheet1のデータからデータフレーム(d1,d2)を作成
df1 = pd.read_excel('2022_年間売上表.xlsx', sheet_name='Sheet1')
df2 = pd.read_excel('2023_年間売上表.xlsx', sheet_name='Sheet1')

#「2022_年間売上表.xlsx」(d1)と「2023_年間売上表.xlsx」(d2)のデータを結合
df = pd.concat([df1, df2], ignore_index=True)

#「商品」「売上年」ごとにグループ化し、合計金額に集約
grouped = df.groupby(['商品', '売上年'], as_index=False)['金額（千円）'].sum()

#エクセルファイル「売上集計表.xlsx」を作成する
output_file = pd.ExcelWriter('売上集計表.xlsx')

#groupedオブジェクトを「売上集計表.xlsx」に書き込み
grouped.to_excel(output_file, sheet_name='売上集計', index=False)

# Excelファイルを保存する
output_file.close()

# 既存ファイルを読み込み
workbook = openpyxl.load_workbook('売上集計表.xlsx')

# アクティブなシートを取得する
ws = workbook.active

# 塗りつぶしスタイルの定義（薄いグレー）
fill = PatternFill(patternType='solid', fgColor='F2F2F2')

# A1〜C1の各セルに塗りつぶしを適用
for cell in ws[1][:3]:
    cell.fill = fill

# 保存
workbook.save('売上集計表.xlsx')
```
