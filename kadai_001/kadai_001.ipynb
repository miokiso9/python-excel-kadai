```python
import openpyxl

# datetimeモジュールのdatetimeクラスをインポートする
from datetime import datetime

#新規エクセルファイルの作成
workbook = openpyxl.Workbook()

# シートの選択
sheet = workbook.active

# セルB2に文書名（請求書）を設定
sheet['B2'].value = '請求書'

# セルB4～B7に会社名、住所、電話番号、担当者名を設定
sheet['B4'].value = '株式会社ABC'
sheet['B5'].value = '〒101-0022 東京都千代田区神田練塀町300'
sheet['B6'].value = 'TEL:03-1234-5678 FAX:03-1234-5678'
sheet['B7'].value = '担当者名:鈴木一郎 様'

# セルF4～G4に請求書番号を設定
sheet['F4'].value = 'No.'
sheet['G4'].value = '0001'

# セルF5～G5に現在日付を設定
sheet['F5'].value = '日付'
today = datetime.today()                          #現在日付を取得
sheet['G5'].value = today.strftime('%Y/%m/%d')    #日付の型を変更して、セルG5に設定

# リストを用いて、商品データを生成する
items = [
    {"商品名": "商品A", "数量": 2, "単価": 10000},  #商品A
    {"商品名": "商品B", "数量": 1, "単価": 15000},  #商品B
]

# 金額を計算してリストに追加
for item in items:
    item["金額"] = item["数量"] * item["単価"]

# ヘッダーをセルB10～E10に出力
header = ["商品名", "数量", "単価", "金額"]                 #ヘッダーの要素を定義
for i, value in enumerate(header):                          #enumerate関数で、ヘッダーのインデックス、値を呼び出し、値を設定
    sheet.cell(row=10, column=2 + i).value = value

 # 11～12行目に商品名・数量・単価・金額を列ごとに出力
for i, item in enumerate(items):                            #enumerate関数で、商品リストのインテックス、値を呼び出し、値を設定
    sheet.cell(row=11 + i, column=2, value=item["商品名"])  # B列
    sheet.cell(row=11 + i, column=3, value=item["数量"])    # C列
    sheet.cell(row=11 + i, column=4, value=item["単価"])    # D列
    sheet.cell(row=11 + i, column=5, value=item["金額"])    # E列

#小計を計算
total_amount = sum(item["金額"] for item in items)
#消費税を計算
tax_amount = total_amount * 0.1
#合計（消費税込み）を計算
total_price = total_amount + tax_amount

#小計を出力
sheet['E13'].value = total_amount
sheet['B15'].value = '小計'
sheet['E15'].value = total_amount

#消費税を出力
sheet['B16'].value = '消費税'
sheet['E16'].value = tax_amount

#合計（消費税込み）を出力
sheet['B17'].value = '合計'
sheet['E17'].value = total_price

# 現在の日付の名前をつけてファイルを保存する
backup_file_name = '請求書_' + today.strftime('%Y%m%d') + '.xlsx'
workbook.save(backup_file_name)

```
